```python
# Complete the functions defined in this cell

# Test 01: Play Right
def rightrotate(arr):   
    last = arr[len(arr)-1];    
    for j in range(len(arr)-1, -1, -1):       
        arr[j] = arr[j-1];       
    arr[0] = last;  
      
def playRight(arr,beats):
    for i in beats:
        if i==1:
            rightrotate(arr)
    return arr


# Test 02: Discard Cards
def leftshift(card):
  start = 0
  end = len(card)
  shift = True
  while(shift == True):
    for i in range(len(card)):
        if card[i] == 0:
            start = i
            break
    for j in range(len(card)-1,-1,-1):
        if card[j] != 0:
            end = j
            break
  
    while(start < end):
        card[start] = card[start+1]
        start += 1
    card[start] = 0
    
    if end < start:
        shift = False
    
  return card

def discardCards(card, number):
  for i in range(len(card)):
    if card[i] == number:
      card[i] = 0
  leftshift(card)
  return card


#Test 03: Merge Lineup
def mergeLineup(pokemon_1, pokemon_2):
  merge = np.array([0]*len(pokemon_1))
  m,n = 0,0
  sum = 0

  for i in range(len(pokemon_1)):
    m = pokemon_1[i]
    n = pokemon_2[len(pokemon_1)-1-i]
    if m!= None and n!= None:
      sum = m + n
    else:
      if m == None:
        if n!= None:
          m = 0
          sum = m + n
        else:
          m,n = 0,0
          sum = m + n
      elif n == None:
        if m!= None:
          n = 0
          sum = m + n
        else:
          m,n = 0,0
          sum = m + n
      else:
        m = 0
        n = 0
        sum = m + n

    merge[i]=sum
  return merge


# Test 04: Balance your Salami
def balanceSalami(salami):
  balance = False
  for i in range(len(salami)):
    sum1 = 0
    sum2 = 0
    for j in range(0,i+1):
      sum1 += salami[j]
    for k in range(i+1,len(salami)):
      sum2 += salami[k]
    if sum1 == sum2:
      balance = True
      break
    else:
      balance = False
  return balance

# Test 05: Protecc Salami
def protectSalami(salami):
  new = np.array([0]*len(salami))
  idx = 0
  for i in salami:
    if i not in new:
      new[idx] = i
      idx += 1

  size = 0
  for i in new:
    if i != 0:
      size += 1
  count = np.array([0]*size)

  for i in range(len(count)):
    c = 0
    for j in range(len(salami)):
      if new[i] == salami[j]:
        c += 1
    count[i] = c

  protect = False
  same = [0]*len(count)
  same[0] = count[0]

  for i in range(1,len(count)):
    if same[0] == count[i]:
      if i >= 2:
        protect = True

  return protect


# Test 05: Odd Even Wave
def waveYourFlag(arr):
  size = 0
  for i in arr:
    if i != 0:
      size += 1

  if size % 2 == 0:
    even = np.array([0]*(size//2))
    odd = np.array([0]*(size//2))
  else:
    even = np.array([0]*(size//2 + 1))
    odd = np.array([0]*(size//2 + 1))
    
  c = 0
  temp = 0
  for i in range(size):
    if arr[i] % 2 == 0:
      even[c] = arr[i]
      c += 1
    else:
      odd[temp] = arr[i]
      temp += 1
  
  new = np.array([0]*size)
  if arr[0] % 2 == 0:
    idx = 0
    for i in range(0,len(new),2):
      new[i] = even[idx]
      idx += 1
    idx = 0
    for j in range(1,len(new),2):
      new[j] = odd[idx]
      idx += 1
  else:
    idx = 0
    for i in range(0,len(new),2):
      new[i] = odd[idx]
      idx += 1
    idx = 0
    for j in range(1,len(new),2):
      new[j] = even[idx]
      idx += 1
  
  return new




```

```python
import math 
import numpy as np 

def mean(a):
  sum = 0
  for x in a:
    sum += x
  return sum/(len(a))

def SD(a):
  m = mean(a)
  s = 0
  l = len(a)
  for x in a:
    s += (x-m)**2
  return round(math.sqrt(s/(l-1)),2)

def newArr(a):
  m = mean(a)
  sd = SD(a)
  count = 0
  for x in a:
    sd_away = (x-m)/sd
    if abs(sd_away) >= 1.5:
      count +=1
  
  if count == 0:
    return None
    
  newA = np.array([0]*count)
  idx = 0
  for x in a:
    sd_away = (x-m)/sd
    if abs(sd_away) >= 1.5:
      newA[idx] = x
      idx += 1

  return newA

a = np.array([10,8,13,9,14,25,-5,20,7,7,4])
print("The mean of the numbers is: ",mean(a))
print("The standard deviation is: ",SD(a))
print("New array: ",newArr(a))
```
